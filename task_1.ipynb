```python
import pandas as pd
```

```python
#loading the file
df = pd.read_csv("marketing_campaign.csv", sep='\t')
display(df)
```

```python
#info
df.info()
```

```python
#checking for missing values
print(df.isnull().sum())
```

```python
#filling missing value in income column using median
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
print(df.isnull().sum())
```

```python
#checking for duplicate values
print(df.duplicated().sum())
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
#standardize Text Columns
text_cols = ['education', 'marital_status']
for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
#handling outliers
q95 = df['income'].quantile(0.95)
df['income'] = df['income'].apply(lambda x: min(x, q95))
```

```python
df.columns
```

```python
#calculating customer age
from datetime import datetime
df['customer_age'] = datetime.now().year - df['year_birth']
```

```python
#calculateing customer tenure
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y')
df['customer_tenure_days'] = (datetime.now() - df['dt_customer']).dt.days
```

```python
df = df[df['customer_age'].between(18, 100)]
```

```python
df.info()
```

```python
df['marital_status'] = df['marital_status'].replace({
    'divorced': 'not married',
    'single': 'not married',
    'together': 'married',
    'married': 'married',
    'widow': 'not married',
    'yolo': 'unknown',
    'absurd': 'unknown'
})
```

```python
#drop Unnecessary Columns
cols_to_drop = ['id', 'z_costcontact', 'z_revenue','year_birth','dt_customer']
df.drop(columns=cols_to_drop, inplace=True)
```
